### 1. In how many cities Summer Olympics is held so far?

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd

df = pd.read_csv('summer.csv')
df_gold = df[df['Medal']=='Gold']

data=[]
for sport in df_gold['Sport'].unique():
    data.append([sport, len(df_gold[df_gold['Sport']==sport])])
data = pd.DataFrame(data, columns=['Sport', 'freq']).sort_values(by='freq', ascending = False).head()
data.plot(x='Sport',y='freq',kind='bar',figsize=(10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []
df = pd.read_csv('summer.csv')
for medal in df['Sport'].unique():
    data.append([medal, len(df[df['Sport'] == medal])])
data = pd.DataFrame(data, columns=['Sport','freq']).sort_values(by='freq',ascending=False).head()
data.plot(x='Sport',y='freq',kind='bar',figsize=(10,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []
for player in df['Athlete'].unique():
    data.append([player, len(df[df['Athlete']==player])])
data = pd.DataFrame(data, columns=['Athlete', 'freq']).sort_values(by='freq',ascending=False).head()
data.plot(x='Athlete', y='freq', kind='bar', figsize=(10,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df = pd.read_csv('summer.csv')
data = []
for event in df['Event'].unique():
    data.append([event, len(df[df['Event'] == event])])
data = pd.DataFrame(data, columns=['Event', 'freq']).sort_values(by='freq', ascending=False).head()
data.plot(x='Event',y='freq', kind='bar', figsize=(10, 5))
```
